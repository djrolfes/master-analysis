```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
betas = np.array([2.0, 2.2, 2.4, 2.5, 2.6, 2.7])
t0s = np.array([0.793, 1.211, 2.814, 5.02, 10.3, 30.0])
dt0s = np.array([0.005, 0.005,0.06,0.17, 0.5, 2])

plt.errorbar(betas[1:], t0s[1:], yerr=dt0s[1:], fmt='.', label='Data with error bars')
plt.grid()
plt.yscale('log')
plt.xlabel('Beta')
plt.ylabel('t0')
plt.title('t0 vs Beta with Error Bars')
```

```python
import scipy.optimize as so
betas = np.array([2.4, 2.5, 2.6])
t0s = np.array([2.815, 5.00, 9.31])
dt0s = np.array([0.008, 0.03,0.08])

def fit_func(beta, a, b):
    return a * np.exp(b * beta)


popt, pcov = so.curve_fit(fit_func, betas, t0s, sigma=dt0s, absolute_sigma=True)
errors = np.sqrt(np.diag(pcov))
a_fit, b_fit = popt
beta_fit = np.linspace(2.3, 2.8, 100)
t0_fit = fit_func(beta_fit, a_fit, b_fit)
plt.errorbar(betas, t0s, yerr=dt0s, fmt='.', label='Data with error bars')
plt.plot(beta_fit, t0_fit, label='Fitted curve', color='red')
plt.grid()
plt.yscale('log')
plt.xlabel('Beta')
plt.ylabel('t0')
plt.title('t0 vs Beta with Error Bars')
plt.xlim(2.3, 2.8)
plt.ylim(2, 15)

print(f"Fitted parameters: a = {a_fit}, b = {b_fit}")
print(f"Covariance matrix:\n{pcov}")
print(f"Parameter errors: {errors}")

print(f"t0(2.8) = {fit_func(2.8, *popt)} +- {np.sqrt(np.diag(pcov))[1] * fit_func(2.8, *popt)}")
```
